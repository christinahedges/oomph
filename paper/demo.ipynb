```python
import oomph
import numpy as np
import astropy.units as u
```

```python
c = oomph.CoRoTMultiPhotometry.from_archive(id=221647275, teff=3984)
```

```python
c
```

```python
# Initial calibration, using parameters with wide range
c.calibrate(p1_lim=(400, 600), p2_lim=(500, 700), plot=False)
# Second calibration, refining initial guess
c.calibrate(p1_lim=(c.p1 - 30, c.p1 + 30), p2_lim=(c.p2 - 30, c.p2 + 30), plot=True);
```

```python
# I've identified the flares in this target by eye
flare_times =[54754.191942697995, 54763.32323887, 54778.85323887, 54781.59323887]
```

```python
for t in flare_times:
    k = (c.lc.time.jd > (t - 0.1)) & (c.lc.time.jd < (t + 0.2))
    c.lc[k].plot()
```

```python
# Iterate through all the flares, and fit them
for t in flare_times:
    k = (c.lc.time.jd > (t - 0.1)) & (c.lc.time.jd < (t + 0.2))
    c.fit_flare(k, r_star=0.76*u.solRad)
```
